```python
# What this script does:
# - Loads /home/ahallur1/spear/Seamless_Experiments/Master/seamless_features.csv
# - Applies consistent "usable iff" filtering:
#     (status == OK) AND (required metric(s) non-null) AND (required strat var present)
# - Generates ONLY prosodic + temporal figures (NO lexical, NO relationship/detail plots)
# - Saves each figure as high-res PNG + PDF
# - Prints a paper-ready caption for EVERY figure (with bin edges in captions where needed)
#
# Figures produced:
#  Fig1 (double-column): Prosody × Gender (3 panels: Mean F0, F0 SD, F0 Range; KDE lines; no grid)
#  Fig2 (single-column): Prosody × Arousal (2 panels: F0 SD + F0 Range; sextile bins; x labels S1..S6; no fliers)
#  Fig3 (single-column): Prosody × Dominance (same as Fig2)
#  Fig4 (double-column): Temporal continuous trends (2 panels:
#       left: Speech rate vs arousal & dominance (two smooth lines + CI)
#       right: Pause ratio vs arousal & dominance (two smooth lines + CI)
#  Fig5 (double-column): Age effects (2 panels:
#       left: Mean F0 by age bin × gender (boxplots; no fliers)
#       right: (stacked mini-panels) speech rate by age bin + pause ratio by age bin (no fliers)
#
# Notes:
# - We DO NOT clip values (no 1..99 clipping, no arbitrary trimming).
# - Boxplots never show outliers (showfliers=False).
# - Continuous trend plots use binned summaries + smoothing to avoid end-of-range instability.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pathlib import Path
from scipy.stats import gaussian_kde

# -------------------------
# Paths / IO
# -------------------------
DATA_PATH = Path("/home/ahallur1/spear/Seamless_Experiments/Master/seamless_features.csv")
OUT_DIR = Path("/home/ahallur1/spear/Seamless_Experiments/Master/master_figures_v5")
OUT_DIR.mkdir(parents=True, exist_ok=True)

print("Loading:", DATA_PATH)
df = pd.read_csv(DATA_PATH)
print("Loaded df:", df.shape)

# -------------------------
# Styling
# -------------------------
sns.set(style="white", context="paper", font_scale=1.25)  # white background, no default grid
# Make axis ticks clearly visible everywhere
plt.rcParams.update({
    "font.family": "serif",
    "font.serif": ["Times New Roman", "Times", "DejaVu Serif"],
    "pdf.fonttype": 42,
    "ps.fonttype": 42,
    "xtick.bottom": True,
    "ytick.left": True,
    "xtick.direction": "out",
    "ytick.direction": "out",
    "xtick.major.size": 6,
    "ytick.major.size": 6,
    "xtick.major.width": 1.2,
    "ytick.major.width": 1.2,
})

def save_fig(fig, stem: str):
    png = OUT_DIR / f"{stem}.png"
    pdf = OUT_DIR / f"{stem}.pdf"
    fig.savefig(png, dpi=1200, bbox_inches="tight")
    fig.savefig(pdf, dpi=1200, bbox_inches="tight")
    print("Saved:", png, "and", pdf)

# -------------------------
# Helpers: normalization
# -------------------------
def normalize_gender(s):
    if not isinstance(s, str):
        return np.nan
    t = s.strip().lower()
    if t in ["male", "m"]:
        return "Male"
    if t in ["female", "f"]:
        return "Female"
    return np.nan

def normalize_status(s):
    if not isinstance(s, str):
        return np.nan
    return s.strip().upper()

def to_num(series):
    return pd.to_numeric(series, errors="coerce")

# -------------------------
# Helper: "usable iff" filtering
# -------------------------
def usable_mask(df_in: pd.DataFrame, status_col: str, required_cols: list, strat_cols: list | None = None):
    """
    usable iff:
      - status_col == "OK"
      - all required_cols are finite
      - all strat_cols (if provided) are non-null (notna)
    """
    m = pd.Series(True, index=df_in.index)

    if status_col in df_in.columns:
        m &= (df_in[status_col] == "OK")
    else:
        # If status col missing, nothing is usable (keeps logic explicit)
        m &= False

    for c in required_cols:
        if c not in df_in.columns:
            m &= False
        else:
            m &= np.isfinite(df_in[c])

    if strat_cols:
        for s in strat_cols:
            if s not in df_in.columns:
                m &= False
            else:
                m &= df_in[s].notna()

    return m

# -------------------------
# REPORTING UTILITIES (add once)
# -------------------------
def _finite(arr):
    arr = np.asarray(arr)
    return arr[np.isfinite(arr)]

def pick_duration_col(df_in: pd.DataFrame):
    """
    Prefer total_duration_s_x (audio duration from prosody merge),
    then total_duration_s_y, then total_duration_s.
    """
    for c in ["total_duration_s_x", "total_duration_s_y", "total_duration_s"]:
        if c in df_in.columns:
            return c
    return None

def report_hours(df_subset: pd.DataFrame, label: str):
    """
    Reports N and total duration in hours for the *exact subset used*.
    """
    dur_col = pick_duration_col(df_subset)
    if dur_col is None:
        print(f"{label}: N={df_subset.shape[0]:,} | duration_col=None | total_hours=NA")
        return
    d = _finite(pd.to_numeric(df_subset[dur_col], errors="coerce").to_numpy())
    total_sec = float(d.sum()) if d.size else 0.0
    print(f"{label}: N={df_subset.shape[0]:,} | duration_col={dur_col} | total_hours={total_sec/3600.0:.3f}")

def report_metric_basic(name: str, x_all: np.ndarray, xlim=None):
    """
    Compact numeric summary for a metric array.
    """
    x = _finite(x_all)
    if x.size == 0:
        print(f"{name}: N=0")
        return
    q25, q50, q75 = np.percentile(x, [25, 50, 75])
    msg = (
        f"{name}: N={x.size:,} | mean={x.mean():.4g} | median={q50:.4g} | "
        f"IQR=[{q25:.4g}, {q75:.4g}]"
    )
    if xlim is not None:
        msg += f" | xlim={xlim}"
    print(msg)

def report_two_groups(gA: str, xA: np.ndarray, gB: str, xB: np.ndarray):
    """
    Compact per-group summary (useful when you draw median lines).
    """
    xa = _finite(xA)
    xb = _finite(xB)
    if xa.size:
        print(f"  {gA}: N={xa.size:,} | mean={xa.mean():.4g} | median={np.median(xa):.4g}")
    else:
        print(f"  {gA}: N=0")
    if xb.size:
        print(f"  {gB}: N={xb.size:,} | mean={xb.mean():.4g} | median={np.median(xb):.4g}")
    else:
        print(f"  {gB}: N=0")

def report_header(title: str):
    print("\n" + "=" * 90)
    print(title)
    print("=" * 90)

# -------------------------
# Helper: KDE as "count-like" curve (matches your current line style)
# -------------------------
def fd_bin_width(x: np.ndarray) -> float:
    """Freedman–Diaconis bin width (used to scale KDE into 'counts')."""
    x = np.asarray(x)
    x = x[np.isfinite(x)]
    n = x.size
    if n < 2:
        return 1.0
    q75, q25 = np.percentile(x, [75, 25])
    iqr = q75 - q25
    if iqr <= 0:
        s = np.std(x)
        if s <= 0:
            return 1.0
        return 3.5 * s / (n ** (1 / 3))
    return 2.0 * iqr / (n ** (1 / 3))

def kde_count_curve(x: np.ndarray, grid: np.ndarray, bin_width: float, bw_method="scott") -> np.ndarray:
    x = np.asarray(x)
    x = x[np.isfinite(x)]
    if x.size < 10:
        return np.full_like(grid, np.nan, dtype=float)
    kde = gaussian_kde(x, bw_method=bw_method)
    return kde(grid) * x.size * bin_width

def robust_xlim(x: np.ndarray, pad_frac: float = 0.03, qlo: float = 0.005, qhi: float = 0.995):
    """
    Avoids useless extreme tails while still covering essentially all data.
    Uses quantiles + padding. Easy to override via MANUAL_XLIMS below.
    """
    x = np.asarray(x)
    x = x[np.isfinite(x)]
    if x.size == 0:
        return (0.0, 1.0)
    lo = float(np.quantile(x, qlo))
    hi = float(np.quantile(x, qhi))
    if hi <= lo:
        lo = float(np.min(x))
        hi = float(np.max(x))
        if hi == lo:
            return (lo - 1.0, hi + 1.0)
    pad = (hi - lo) * pad_frac
    return (lo - pad, hi + pad)

def make_grid(xlim, n=900):
    return np.linspace(xlim[0], xlim[1], n)

def draw_fullheight_vline(ax, x, color, linestyle="--", linewidth=1.8, alpha=0.95):
    ymin, ymax = ax.get_ylim()
    ax.vlines(x, ymin, ymax, colors=color, linestyles=linestyle, linewidth=linewidth, alpha=alpha)

# -------------------------
# Helper: stable continuous trend with error bands
# -------------------------
def binned_trend_with_ci(
    x: np.ndarray,
    y: np.ndarray,
    n_bins: int = 30,
    min_count: int = 200,
    smooth_window: int = 5,
    ci_method: str = "sem",  # "sem" or "bootstrap"
    n_boot: int = 300,
    seed: int = 0,
):
    """
    Returns arrays for a stable trend line and an error band by:
      1) binning x into equal-count bins (quantile bins)
      2) computing bin centers + mean(y) per bin
      3) computing error per bin (SEM or bootstrap CI)
      4) smoothing with a rolling median (helps avoid end spikes)
      5) dropping bins with too few points (min_count)

    This prevents sparse extremes from creating weird bends/peaks.
    """
    rng = np.random.default_rng(seed)
    x = np.asarray(x)
    y = np.asarray(y)
    m = np.isfinite(x) & np.isfinite(y)
    x = x[m]
    y = y[m]
    if x.size < max(min_count, 50):
        return None

    # quantile edges for equal-count bins
    qs = np.linspace(0, 1, n_bins + 1)
    edges = np.quantile(x, qs)
    # ensure strictly increasing edges (handle ties)
    edges = np.unique(edges)
    if edges.size < 6:
        return None

    # assign bins
    bin_idx = np.digitize(x, edges[1:-1], right=True)

    centers = []
    means = []
    lowers = []
    uppers = []
    counts = []

    for b in range(0, edges.size - 1):
        mask_b = (bin_idx == b)
        yb = y[mask_b]
        xb = x[mask_b]
        if yb.size < min_count:
            continue

        c = float(np.median(xb))
        mu = float(np.mean(yb))

        if ci_method == "bootstrap":
            boots = []
            for _ in range(n_boot):
                samp = rng.choice(yb, size=yb.size, replace=True)
                boots.append(np.mean(samp))
            lo, hi = np.quantile(boots, [0.025, 0.975])
        else:
            # SEM-based approx 95% CI
            sem = float(np.std(yb, ddof=1) / np.sqrt(yb.size))
            lo = mu - 1.96 * sem
            hi = mu + 1.96 * sem

        centers.append(c)
        means.append(mu)
        lowers.append(lo)
        uppers.append(hi)
        counts.append(yb.size)

    if len(centers) < 6:
        return None

    centers = np.array(centers)
    means = np.array(means)
    lowers = np.array(lowers)
    uppers = np.array(uppers)
    counts = np.array(counts)

    # sort by x
    order = np.argsort(centers)
    centers, means, lowers, uppers, counts = centers[order], means[order], lowers[order], uppers[order], counts[order]

    # rolling median smoothing (robust to outliers)
    def roll_median(a, w):
        s = pd.Series(a)
        return s.rolling(w, center=True, min_periods=max(2, w//2)).median().to_numpy()

    w = int(max(3, smooth_window))
    means_s = roll_median(means, w)
    lowers_s = roll_median(lowers, w)
    uppers_s = roll_median(uppers, w)

    return dict(
        x=centers,
        y=means_s,
        lo=lowers_s,
        hi=uppers_s,
        n=counts
    )

# -------------------------
# Columns we use (prosody + temporal only)
# -------------------------
PROSODY_METRICS = ["f0_mean_p10_p90", "f0_std_p10_p90", "f0_range_p10_p90"]
TEMP_METRICS = ["speech_rate_wpm", "articulation_rate_wpm", "pause_ratio", "pause_mean_duration_s"]

# normalize key columns
df["gender2"] = df["gender"].apply(normalize_gender) if "gender" in df.columns else np.nan
df["status_x"] = df["status_x"].apply(normalize_status) if "status_x" in df.columns else np.nan
df["status_y"] = df["status_y"].apply(normalize_status) if "status_y" in df.columns else np.nan

for c in PROSODY_METRICS + TEMP_METRICS + ["arousal", "dominance", "age_years"]:
    if c in df.columns:
        df[c] = to_num(df[c])

# age bins (ignore <18)
def age_bin(a):
    if not np.isfinite(a):
        return np.nan
    if a < 18:
        return np.nan
    if a <= 29:
        return "18-29"
    if a <= 59:
        return "30-59"
    return "60+"

df["age_bin"] = df["age_years"].apply(age_bin) if "age_years" in df.columns else np.nan
AGE_ORDER = ["18-29", "30-59", "60+"]
df["age_bin"] = pd.Categorical(df["age_bin"], categories=AGE_ORDER, ordered=True)

# -------------------------
# Manual xlims (easy to edit)
# -------------------------
MANUAL_XLIMS = {
    # Prosody (leave None unless you want to force)
    "f0_mean_p10_p90": None,
    "f0_std_p10_p90": None,
    "f0_range_p10_p90": None,

    # Temporal
    "speech_rate_wpm": None,
    "articulation_rate_wpm": None,
    "pause_ratio": None,  # per your preference
    "pause_mean_duration_s": None,  # keep full but robust tails via robust_xlim
}

# =============================================================================
# FIGURE 1 — Prosody × Gender (3 panels; KDE lines; no grids)
# =============================================================================
# usable iff:
#  - status_x == OK
#  - required metrics present
#  - gender present

mask_fig1 = usable_mask(df, "status_x", required_cols=PROSODY_METRICS, strat_cols=["gender2"])
df1 = df.loc[mask_fig1 & df["gender2"].isin(["Male", "Female"])].copy()
print("\nFig1 usable rows:", df1.shape)

report_header("FIGURE 1 — Prosody × Gender (usable subset)")
report_hours(df1, "Fig1 subset")

fig = plt.figure(figsize=(13.5, 4.2))  # double-column
for i, metric in enumerate(PROSODY_METRICS):

    ax = plt.subplot(1, 3, i + 1)
    ax.grid(False)

    x_m = df1.loc[df1["gender2"] == "Male", metric].to_numpy()
    x_f = df1.loc[df1["gender2"] == "Female", metric].to_numpy()
    x_all = df1[metric].to_numpy()

    xlim = MANUAL_XLIMS.get(metric) or robust_xlim(x_all, qlo=0.003, qhi=0.997)
    grid = make_grid(xlim, n=900)
    bw = fd_bin_width(x_all)

    y_m = kde_count_curve(x_m, grid, bin_width=bw)
    y_f = kde_count_curve(x_f, grid, bin_width=bw)

    line_m = ax.plot(grid, y_m, linewidth=2.4, label="Male")[0]
    line_f = ax.plot(grid, y_f, linewidth=2.4, label="Female")[0]

    ax.set_xlim(*xlim)
    if metric == "f0_mean_p10_p90":
        ax.set_xlabel("F0 Mean (10–90%) [Hz]")
        ax.set_ylabel("Count")
    elif metric == "f0_std_p10_p90":
        ax.set_xlabel("F0 SD (10–90%) [Hz]")
    else:
        ax.set_xlabel("F0 Range (10–90%) [Hz]")

    ax.relim()
    ax.autoscale_view()

    med_m = float(np.nanmedian(x_m[np.isfinite(x_m)]))
    med_f = float(np.nanmedian(x_f[np.isfinite(x_f)]))
    draw_fullheight_vline(ax, med_m, color=line_m.get_color())
    draw_fullheight_vline(ax, med_f, color=line_f.get_color())

    if i == 0:
        ax.legend(frameon=True)

    report_metric_basic(metric, x_all, xlim)
    report_two_groups("Male", x_m, "Female", x_f)

plt.tight_layout()
save_fig(fig, "Fig1_Prosody_Gender")
plt.show()

# Caption for Fig1
cap1 = (
    "Figure 1. Prosodic operating regimes (10–90% trimmed) stratified by gender. "
    "Kernel density curves (scaled to counts) are shown for Mean F0, F0 SD, and F0 Range. "
    "Vertical dashed lines indicate group medians. "
    f"Usable samples required status_x==OK, non-missing prosodic metrics, and a normalized gender label "
    f"(N={len(df1):,}; Male={int((df1['gender2']=='Male').sum()):,}, Female={int((df1['gender2']=='Female').sum()):,})."
)
print("\nCAPTION (Fig1):\n", cap1)

# =============================================================================
# FIGURE 2 — Prosody × Arousal (single-column; sextile bins; no numeric x tick labels)
# =============================================================================
# usable iff:
#  - status_x == OK
#  - required metrics present
#  - arousal present

mask_fig2 = usable_mask(df, "status_x", required_cols=["f0_std_p10_p90", "f0_range_p10_p90", "arousal"], strat_cols=None)
df2 = df.loc[mask_fig2].copy()
print("\nFig2 usable rows:", df2.shape)

# sextile binning (equal-count bins)
df2["arousal_bin"] = pd.qcut(df2["arousal"], q=6, labels=[f"S{i}" for i in range(1, 7)])

# capture bin edges for caption
ar_edges = pd.qcut(df2["arousal"], q=6, retbins=True, duplicates="drop")[1]
ar_edges = [float(x) for x in ar_edges]

report_header("FIGURE 2 — Prosody × Arousal (binned) usable subset")
report_hours(df2, "Fig2 subset")
report_metric_basic("Arousal (pooled)", df2["arousal"].to_numpy())
report_metric_basic("F0 SD (10–90%)", df2["f0_std_p10_p90"].to_numpy())
report_metric_basic("F0 Range (10–90%)", df2["f0_range_p10_p90"].to_numpy())
print("Arousal sextile edges:", [round(x, 4) for x in ar_edges])
print("Counts per sextile:\n", df2["arousal_bin"].value_counts().sort_index())

fig = plt.figure(figsize=(7.0, 4.2))  # single-column
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)
for ax in [ax1, ax2]:
    ax.grid(False)

sns.boxplot(
    data=df2, x="arousal_bin", y="f0_std_p10_p90",
    ax=ax1, showfliers=False
)
ax1.set_title("F0 SD v/s Arousal")
ax1.set_xlabel("Arousal Sextiles")
ax1.set_ylabel("F0 SD (10–90%) [Hz]")
ax1.set_xticklabels([f"S{i}" for i in range(1, 7)])

sns.boxplot(
    data=df2, x="arousal_bin", y="f0_range_p10_p90",
    ax=ax2, showfliers=False
)
ax2.set_title("F0 Range v/s Arousal")
ax2.set_xlabel("Arousal Sextiles")
ax2.set_ylabel("F0 Range (10–90%) [Hz]")
ax2.set_xticklabels([f"S{i}" for i in range(1, 7)])

plt.tight_layout()
save_fig(fig, "Fig2_Prosody_Arousal_Bins")
plt.show()

cap2 = (
    "Figure 2. State-driven pitch expressivity across arousal sextiles (S1–S6). "
    "Boxplots summarize F0 SD and F0 Range (10–90% trimmed) within equal-count arousal bins "
    "(outliers not shown). "
    "Arousal bin edges: " + ", ".join([f"{ar_edges[i]:.3f}–{ar_edges[i+1]:.3f}" for i in range(len(ar_edges)-1)]) + ". "
    f"Usable samples required status_x==OK and non-missing arousal + prosodic metrics "
    f"(N={len(df2):,})."
)
print("\nCAPTION (Fig2):\n", cap2)

# =============================================================================
# FIGURE 3 — Prosody × Dominance (single-column; sextile bins; no numeric x tick labels)
# =============================================================================
mask_fig3 = usable_mask(df, "status_x", required_cols=["f0_std_p10_p90", "f0_range_p10_p90", "dominance"], strat_cols=None)
df3 = df.loc[mask_fig3].copy()
print("\nFig3 usable rows:", df3.shape)

df3["dom_bin"] = pd.qcut(df3["dominance"], q=6, labels=[f"S{i}" for i in range(1, 7)])
dom_edges = pd.qcut(df3["dominance"], q=6, retbins=True, duplicates="drop")[1]
dom_edges = [float(x) for x in dom_edges]

report_header("FIGURE 3 — Prosody × Dominance (binned) usable subset")
report_hours(df3, "Fig3 subset")
report_metric_basic("Dominance (pooled)", df3["dominance"].to_numpy())
report_metric_basic("F0 SD (10–90%)", df3["f0_std_p10_p90"].to_numpy())
report_metric_basic("F0 Range (10–90%)", df3["f0_range_p10_p90"].to_numpy())
print("Dominance sextile edges:", [round(x, 4) for x in dom_edges])
print("Counts per sextile:\n", df3["dom_bin"].value_counts().sort_index())

fig = plt.figure(figsize=(7.0, 4.2))  # single-column
ax1 = plt.subplot(1, 2, 1)
ax2 = plt.subplot(1, 2, 2)
for ax in [ax1, ax2]:
    ax.grid(False)

sns.boxplot(
    data=df3, x="dom_bin", y="f0_std_p10_p90",
    ax=ax1, showfliers=False
)
ax1.set_title("F0 SD v/s Dominance")
ax1.set_xlabel("Dominance Sextiles")
ax1.set_ylabel("F0 SD (10–90%) [Hz]")
ax1.set_xticklabels([f"S{i}" for i in range(1, 7)])

sns.boxplot(
    data=df3, x="dom_bin", y="f0_range_p10_p90",
    ax=ax2, showfliers=False
)
ax2.set_title("F0 Range v/s Dominance")
ax2.set_xlabel("Dominance Sextiles")
ax2.set_ylabel("F0 Range (10–90%) [Hz]")
ax2.set_xticklabels([f"S{i}" for i in range(1, 7)])

plt.tight_layout()
save_fig(fig, "Fig3_Prosody_Dominance_Bins")
plt.show()

cap3 = (
    "Figure 3. Pitch expressivity across dominance sextiles (S1–S6). "
    "Boxplots summarize F0 SD and F0 Range (10–90% trimmed) within equal-count dominance bins "
    "(outliers not shown). "
    "Dominance bin edges: " + ", ".join([f"{dom_edges[i]:.3f}–{dom_edges[i+1]:.3f}" for i in range(len(dom_edges)-1)]) + ". "
    f"Usable samples required status_x==OK and non-missing dominance + prosodic metrics "
    f"(N={len(df3):,})."
)
print("\nCAPTION (Fig3):\n", cap3)

# =============================================================================
# FIGURE 4 — Temporal continuous trends (double-column): speech_rate & pause_ratio vs arousal/dominance
# =============================================================================
# usable iff:
#  - status_y == OK
#  - required temporal metrics present
#  - arousal/dominance present
mask_fig4 = usable_mask(
    df, "status_y",
    required_cols=["speech_rate_wpm", "pause_ratio", "arousal", "dominance"],
    strat_cols=None
)
df4 = df.loc[mask_fig4].copy()
print("\nFig4 usable rows:", df4.shape)

report_header("FIGURE 4 — Temporal trends v/s state (usable subset)")
report_hours(df4, "Fig4 subset")
report_metric_basic("Speech rate (WPM)", df4["speech_rate_wpm"].to_numpy())
report_metric_basic("Pause ratio", df4["pause_ratio"].to_numpy())
report_metric_basic("Arousal", df4["arousal"].to_numpy())
report_metric_basic("Dominance", df4["dominance"].to_numpy())

# Build stable trends
trend_sr_ar = binned_trend_with_ci(df4["arousal"].to_numpy(), df4["speech_rate_wpm"].to_numpy(),
                                  n_bins=30, min_count=200, smooth_window=5, ci_method="sem")
trend_sr_dom = binned_trend_with_ci(df4["dominance"].to_numpy(), df4["speech_rate_wpm"].to_numpy(),
                                   n_bins=30, min_count=200, smooth_window=5, ci_method="sem")

trend_pr_ar = binned_trend_with_ci(df4["arousal"].to_numpy(), df4["pause_ratio"].to_numpy(),
                                  n_bins=30, min_count=200, smooth_window=5, ci_method="sem")
trend_pr_dom = binned_trend_with_ci(df4["dominance"].to_numpy(), df4["pause_ratio"].to_numpy(),
                                   n_bins=30, min_count=200, smooth_window=5, ci_method="sem")

fig = plt.figure(figsize=(13.5, 4.6))  # double-column, 2 panels
axL = plt.subplot(1, 2, 1)
axR = plt.subplot(1, 2, 2)
axL.grid(False)
axR.grid(False)

# Left panel: Speech rate trend lines + CI
axL.set_title("Speech Rate v/s Arousal & Dominance")
axL.set_xlabel("State Value")
axL.set_ylabel("Speech Rate (WPM)")
axL.set_xlim(0.2, 0.8)

if trend_sr_ar is not None:
    line_ar = axL.plot(trend_sr_ar["x"], trend_sr_ar["y"], linewidth=2.6, label="Arousal")[0]
    axL.fill_between(trend_sr_ar["x"], trend_sr_ar["lo"], trend_sr_ar["hi"], alpha=0.18, color=line_ar.get_color())
if trend_sr_dom is not None:
    line_dom = axL.plot(trend_sr_dom["x"], trend_sr_dom["y"], linewidth=2.6, label="Dominance")[0]
    axL.fill_between(trend_sr_dom["x"], trend_sr_dom["lo"], trend_sr_dom["hi"], alpha=0.18, color=line_dom.get_color())

axL.legend(frameon=True)

# Right panel: Pause ratio trend lines + CI
axR.set_title("Pause Ratio v/s Arousal & Dominance")
axR.set_xlabel("State Value")
axR.set_ylabel("Pause Ratio")
axR.set_xlim(0.2, 0.8)

if trend_pr_ar is not None:
    line_ar = axR.plot(trend_pr_ar["x"], trend_pr_ar["y"], linewidth=2.6, label="Arousal")[0]
    axR.fill_between(trend_pr_ar["x"], trend_pr_ar["lo"], trend_pr_ar["hi"], alpha=0.18, color=line_ar.get_color())
if trend_pr_dom is not None:
    line_dom = axR.plot(trend_pr_dom["x"], trend_pr_dom["y"], linewidth=2.6, label="Dominance")[0]
    axR.fill_between(trend_pr_dom["x"], trend_pr_dom["lo"], trend_pr_dom["hi"], alpha=0.18, color=line_dom.get_color())

axR.legend(frameon=True)

plt.tight_layout()
save_fig(fig, "Fig4_Temporal_Trends_State")
plt.show()

cap4 = (
    "Figure 4. Conversational rhythm varies systematically with speaker state. "
    "Smoothed trend lines show speech rate (left) and pause ratio (right) as functions of arousal and dominance "
    "(0–1 scale). Trends are computed using equal-count binning over the state axis, "
    "followed by robust smoothing; shaded bands indicate approximate 95% uncertainty (SEM-based). "
    "Bins with sparse data are excluded to prevent unstable end effects. "
    f"Usable samples required status_y==OK and non-missing speech_rate_wpm, pause_ratio, arousal, and dominance "
    f"(N={len(df4):,})."
)
print("\nCAPTION (Fig4):\n", cap4)

# =============================================================================
# FIGURE 5 — Age effects (double-column):
#   Left: Mean F0 by age bin × gender (boxplots; no fliers)
#   Right: speech rate by age bin + pause ratio by age bin (stacked mini-panels; no fliers)
# =============================================================================
# Left panel uses prosody status_x OK and needs: f0_mean_p10_p90, age_bin, gender
mask_fig5_left = usable_mask(
    df, "status_x",
    required_cols=["f0_mean_p10_p90"],
    strat_cols=["age_bin", "gender2"]
)
df5L = df.loc[mask_fig5_left & df["gender2"].isin(["Male", "Female"])].copy()
print("\nFig5-left usable rows:", df5L.shape)

# Right panel uses temporal status_y OK and needs: speech_rate_wpm, pause_ratio, age_bin
mask_fig5_right = usable_mask(
    df, "status_y",
    required_cols=["speech_rate_wpm", "pause_ratio"],
    strat_cols=["age_bin"]
)
df5R = df.loc[mask_fig5_right].copy()
print("Fig5-right usable rows:", df5R.shape)

report_header("FIGURE 5 — Age effects (usable subsets)")
report_hours(df5L, "Fig5 LEFT subset (Mean F0 by age×gender)")
print("Age bins (left):\n", df5L["age_bin"].value_counts().sort_index())
print("Gender (left):\n", df5L["gender2"].value_counts().sort_index())
report_metric_basic("Mean F0 (10–90%)", df5L["f0_mean_p10_p90"].to_numpy())

report_hours(df5R, "Fig5 RIGHT subset (Temporal by age)")
print("Age bins (right):\n", df5R["age_bin"].value_counts().sort_index())
report_metric_basic("Speech rate (WPM)", df5R["speech_rate_wpm"].to_numpy())
report_metric_basic("Pause ratio", df5R["pause_ratio"].to_numpy())

fig = plt.figure(figsize=(13.5, 4.8))  # double-column

# ---- Left: Mean F0 by Age × Gender
axL = plt.subplot(1, 2, 1)
axL.grid(False)

sns.boxplot(
    data=df5L,
    x="age_bin",
    y="f0_mean_p10_p90",
    hue="gender2",
    showfliers=False,
    ax=axL
)
axL.set_title("Mean F0 by Age Group and Gender")
axL.set_xlabel("Age Group")
axL.set_ylabel("Mean F0 (10–90%) [Hz]")
axL.legend(title="Gender", frameon=True)

# ---- Right: two stacked mini-panels (speech rate + pause ratio by age bin)
gs = fig.add_gridspec(1, 2, width_ratios=[1.1, 1.0])
sub = gs[0, 1].subgridspec(2, 1, hspace=0.35)

axR1 = fig.add_subplot(sub[0, 0])
axR2 = fig.add_subplot(sub[1, 0])
axR1.grid(False)
axR2.grid(False)

sns.boxplot(
    data=df5R,
    x="age_bin",
    y="speech_rate_wpm",
    showfliers=False,
    ax=axR1
)
axR1.set_title("Speech Rate by Age Group")
axR1.set_xlabel("")
axR1.set_ylabel("Speech Rate (WPM)")

sns.boxplot(
    data=df5R,
    x="age_bin",
    y="pause_ratio",
    showfliers=False,
    ax=axR2
)
axR2.set_title("Pause Ratio by Age Group")
axR2.set_xlabel("Age Group")
axR2.set_ylabel("Pause ratio")

plt.tight_layout()
save_fig(fig, "Fig5_Age_Prosody_Temporal")
plt.show()

cap5 = (
    "Figure 5. Age-group effects on prosody and conversational timing. "
    "(Left) Mean F0 (10–90% trimmed) by Vox-Profile age bins (18–29 / 30–59 / 60+) stratified by gender "
    "(outliers not shown). "
    "(Right) Speech rate and pause ratio distributions by the same age bins (outliers not shown). "
    "Age binning follows Vox-Profile guidance and excludes predicted ages <18 as unreliable. "
    f"Left panel usable samples required status_x==OK, non-missing Mean F0, age bin, and gender (N={len(df5L):,}). "
    f"Right panel usable samples required status_y==OK, non-missing temporal metrics, and age bin (N={len(df5R):,})."
)
print("\nCAPTION (Fig5):\n", cap5)

print("\nAll figures written to:", OUT_DIR)
```
